# 지속가능경영보고서 PDF를 docx로 변경 후, CSV로 변경하는 notebook

#### 기존에 있는 라이브러리가 pdf의 텍스트를 얼마나 잘 인식하는지 확인하고자 우선 docx로 변환
#### docx를 csv로 변경하면서 카탈로그 형태의 텍스트가 csv로 변환됨에 따라 어느 부분에 손실이 발생하는지 확인하고자 했음

## '22년 8월 29일 기준 결과

### PDF -> docx 변환
* 사용한 라이브러리: pdf2docx
* 대부분의 텍스트를 잘 인식함
* 표 같은 경우도 텍스트로 잘 바꿔줬음

### docx -> csv 변환
* 사용한 라이브러리: docx2txt
* docx2txt 라이브러리를 사용해서 docx 내용을 str로 추출
* 추출한 str을 한줄 단위로 csv의 한 행씩 작성함
* 문제점:    
(1) 표를 인식 못함. 행과 열 구분 없이 위에서 아래로, 왼쪽에서 오른쪽으로 읽음 (그냥 줄글 읽듯이)    
(2)추출한 str에 한국어 문법상 콤마(,)가 있는 경우, 이를 csv의 seperator로 인식하여 한 문장이 완결되기 전에 문장이 잘려버리며 csv 상 다음 행으로 내용을 넘겨버림

```python
from pdf2docx import Converter
import sys
import re


if len(sys.argv)== 1:
    print ("USAGE: %s file.pdf", sys.argv[0])
print (sys.argv)

pdf_file = "Downloads/KOR_SK Inc_SR_2021_Interactive_Single_Final.pdf"
docx_file = re.sub(".pdf", ".docx", pdf_file, flags=re.I)

# convert pdf to docx
cv = Converter(pdf_file)
cv.convert(docx_file, start=0, end=None)
cv.close()
```

```python
import pandas as pd
import numpy as np
import docx2txt
```

```python
text = docx2txt.process("KOR_SK Inc_SR_2021_Interactive_Single_Final.docx")
dic = {x for x in text.split('\n')}
#remove the empty spaces in dic
dic.remove('') if '' in dic else dic
```

```python
text_line = [x for x in text.split('\n')]
```

```python
while '' in text_line:
    text_line.remove('')
```

```python
f = open("SK_INC_lineByline.csv","w", encoding='utf-8-sig')
```

```python
for i in range(len(text_line)):
    f.write(text_line[i]+ '\n')
```
