```python
# 폴더 생성
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
        else:
            return -2
    except OSError:
        print ('Error: Creating directory. ' +  directory)
```

```python
path = "kr/*.pdf"   # 현재 directory에서 .png로 끝나는 모든 파일명
file_list = glob.glob(path)
file_list

try:
    for i in range(len(file_list)):
        pdf_path = file_list[i]
        company = pdf_path[3:-4] # 회사명 추출
        save_path = "jpgs/" + company
        print("변환 중인 회사 : " + company)

        if createFolder(save_path) == -2: # 이미 존재하는 폴더명이면
            print("이미 존재하는 폴더입니다.")
            continue

        images = convert_from_path(pdf_path) # 이미지 변환

        for j in range(len(images)):
            images[j].save(save_path + '/page_{}.jpg'.format(j+1))

        print(company, "변환 완료하였습니다.")
except:
    print(company, "는 오류가 발생하는 PDF 파일입니다.")
    pass
```

#### CasCadeTabNet 코드

```python
import pandas as pd

folder_path = "jpgs/*"
folder_list= glob.glob(folder_path)

# 실제 폴더명만 추출
temp_list = []
file_list = []
search = "jpgs/"
for word in folder_list:
    if search in word: 
        temp_list.append(word[5:])

folder_name = temp_list

for i in range(len(folder_list)):
        #print("현재/전체 : ", (i + 1) + "/" + len(folder_list))
        file_path = folder_list[i]
        file_list = glob.glob(file_path + "/*")

        save_path = "./jpgs_cascade/" + folder_name[i]
        createFolder(save_path)
        print(folder_name[i], "CasCadeTabNet 변환")
        for j in range(len(file_list)):
            # Run Inference
            result = inference_detector(model, file_list[j])
            show_result_pyplot(model, file_list[j], result, score_thr=0.85, out_file='jpgs_cascade/{}'.format(file_list[j][5:]))
            if int(((j + 1) / len(file_list)) * 100) % 10 == 0:
                print(folder_name[i], "현재/전체 : ", (j + 1), "/", len(file_list))
                
print("CasCade 종료")
```
