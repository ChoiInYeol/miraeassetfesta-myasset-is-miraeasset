```python
from konlpy.tag import Komoran
import kss
import pytextrank
import spacy
from krwordrank.word import KRWordRank
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/KHUDA/1기/동아리 연합공모전/data/2022_ESG.csv')
```

```python
i = 0
E_sentence = []
S_sentence = []
G_sentence = []
while i != 73:
    E_sentence.append(df[['Environment']].iloc[i][0])
    S_sentence.append(df[['Society']].iloc[i][0])
    G_sentence.append(df[['Governance']].iloc[i][0])
    i += 1
E_sentence
```

```python
all_sentence1 = "".join([str(_) for _ in E_sentence])
all_sentence2 = "".join([str(_) for _ in S_sentence])
all_sentence3 = "".join([str(_) for _ in G_sentence])
```

```python
all_sentence1
```

```python
texts = [all_sentence1]
wordrank_extractor = KRWordRank(min_count=5, max_length=10)
E_keywords, E_rank, E_graph = wordrank_extractor.extract(texts, num_keywords=100)
```

```python
texts = [all_sentence2]
wordrank_extractor = KRWordRank(min_count=5, max_length=10)
S_keywords, S_rank, S_graph = wordrank_extractor.extract(texts, num_keywords=100)
```

```python
texts = [all_sentence3]
wordrank_extractor = KRWordRank(min_count=5, max_length=10)
G_keywords, G_rank, G_graph = wordrank_extractor.extract(texts, num_keywords=100)
```
