```python
# COMPAS 시스템에 특허 검색을 하기 위한 검색식(query)를 분야별로 정리한 딕셔너리 자료형
# 분야별로 출원인명(기업명), 출원인 국가, 특허의 text 필드(title, abstract, description, specification, claim)를 저장
total_query = {"Electrical_Electronic_Equipment": "",
"Services": "",
"Transport_Equipment": "",
"Medical_Supplies": "",
"Distribution": "",
"Iron_Metal_Products": "",
"Other_Financial_Companies": "",
"Chemicals": "",
"Electricity_Gas": "",
"Banks": "",
"Communication": "",
"Insurance": "",
"Non_metallic_Mineral_Products": "",
"Transport_Storage": "",
"Machinery": "",
"Construction": "",
"Food_Beverages": "",
"Securities": "",
"Medical_Precision_Machines": "",
"Other_Manufacturing_Industries": "",
"Textile_Wearing_Apparel": "",
"Mining": "",
"Paper_Wood": "",
"Agriculture_Forestry_Fishing": "",
"Digital_Contents": "",
"Electric_Electronics": "",
"Recreation_Culture": "",
"Broadcasting": "",
"Semiconductors": "",
"Pharmaceuticals": "",
"Financials": "",
"Paper_Wood_Products": "",
"Other_Services": "",
"Software": "",
"IT_Components": "",
"Hotels_Restaurants": "",
"Telecom_Broadcasting_Equipment": "",
"Computer_Services": "",
"Food_Beverage_Tobacco": "",
"Non_metal_Products": "",
"Internet": "",
"Metal_Products": "",
"Vehicles_Components": ""}
```

```python
# 업종별 query에 특허 출원인명(기업명) 추가

for industry in range(len(esg_train["영문 업종명"].unique())):
    an = "an:("
    for company in range(len(esg_train)):
        if str(esg_train["영문 업종명"][company]) == str(esg_train["영문 업종명"].unique()[industry]):
            if an == "an:(":
                an +=  str('"' + str(esg_train["영문 종목명"][company]) + '"')
            else:                  
                an += str(" OR " + '"' + str(esg_train["영문 종목명"][company]) + '"')
    total_query[esg_train["영문 업종명"].unique()[industry]] = an
    
for industry in range(len(esg_test["영문 업종명"].unique())):
    an = ""
    for company in range(len(esg_test)):
        if str(esg_test["영문 업종명"][company]) == str(esg_test["영문 업종명"].unique()[industry]):
            an += str(" OR " + '"' + str(esg_test["영문 종목명"][company]) + '"')
    total_query[esg_test["영문 업종명"].unique()[industry]] += an

for industry in industry_list:
    total_query[industry] += ")"
```

```python
# COMPAS의 검색 시스템에서 오류를 발생시키는 회사명 제거
# 검색 시스템의 명령어와 유사한 회사명은 검색 과정 중 오류를 발생시키는 버그인 것으로 파악
total_query['Chemicals'] = total_query['Chemicals'].replace(' OR "SH ENERGY & CHEMICAL"', "")
total_query["Electrical_Electronic_Equipment"] = total_query["Electrical_Electronic_Equipment"].replace(' OR "LS ELECTRIC"', "")
total_query["Other_Financial_Companies"] = total_query["Other_Financial_Companies"].replace(' OR "LS"', "")
total_query["Other_Financial_Companies"] = total_query["Other_Financial_Companies"].replace(' OR "LS C&S ASIA"', "")
```

```python
# 업종별 query에 특허 출원인 국가 추가
# KOSPI, KOSDAQ에 등록된 기업들로 출원인 국가를 우리나라(KO)로 한정

for industry in industry_list:
    total_query[industry] += " AND acn:KR"
```

```python
# E/S/G 각 항목별 키워드
# 키워드는 산업통상자원부, 한국생산성본부에서 제공한 K-ESG 가이드라인 V1.0의 기본 진단항목 정의서의 항목들에서 추출함
# 평가항목은 기본 진단항목 정의서의 항목이며, 영문 평가항목은 이를 영어로 번역, 키워드는 영문 평가항목을 키워드 위주로 추출한 내용
e_keywords = pd.read_excel("e_keywords.xlsx")
s_keywords = pd.read_excel("s_keywords.xlsx")
g_keywords = pd.read_excel("g_keywords.xlsx")
```

```python
import re

# 검색식에서 text 필드의 들어갈 esg 키워드를 담은 리스트
text_list = []

# E/S/G 각 분야의 키워드를 text_list에 추가
for i in range(len(e_keywords['키워드'])):
    words_list = e_keywords['키워드'][i]
    if str(words_list) != "nan":
        words_sep = re.split(r", |,", words_list)
        for word in words_sep:
            text_list.append(word)
        
for i in range(len(s_keywords['키워드'])):
    words_list = s_keywords['키워드'][i]
    if str(words_list) != "nan":
        words_sep = re.split(r", |,", words_list)
        for word in words_sep:
            text_list.append(word)
        
for i in range(len(g_keywords['키워드'])):
    words_list = g_keywords['키워드'][i]
    if str(words_list) != "nan":
        words_sep = re.split(r", |,", words_list)
        for word in words_sep:
            text_list.append(word)
        

# 중복되는 키워드는 제거        
text_list = list(set(text_list))
```

```python
text_query = "text:("
for text in text_list:
    if text_query == "text:(":
        # *을 키워드 뒤에 붙이는 이유는 COMPAS의 검색식에 키워드*로 입력되면, 해당 키워드를 포함하는 모든 단어를 검색 결과로 
        # 보여주는 COMPAS의 규칙때문
        text_query += str('"' + text + '"')
    else:
        text_query += str(' OR "' + text + '"')
text_query += ")"
```

```python
# 업종별 query에 특허 출원인명(기업명) 추가

for industry in range(len(esg_train["영문 업종명"].unique())):
    total_query[esg_train["영문 업종명"].unique()[industry]] += str(" AND "+ text_query)
```

```python
# 업종별 query를 txt로 저장

for industry in range(len(esg_train["영문 업종명"].unique())):
    f = open("query/"+str(esg_train["영문 업종명"].unique()[industry]) + ".txt", 'w')
    f.write(str(total_query[esg_train["영문 업종명"].unique()[industry]]))
    f.close()
```
